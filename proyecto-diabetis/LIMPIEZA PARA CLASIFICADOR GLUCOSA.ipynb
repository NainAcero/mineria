```python
import numpy as np
import pandas as pd
import seaborn as sb
from scipy import stats
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('dataset/diabetes.csv')
df.head()
```

```python
df["PlasmaGlucose"].head()
```

```python
mpg_labels = [1, 2, 3]
cortes=[1,85,110,30000]

bins_PlasmaGlucose = pd.cut(df['PlasmaGlucose'],cortes, labels=mpg_labels)
bins_PlasmaGlucose.head()
```

```python
bins_PlasmaGlucose.value_counts()
```

```python
df['rangos de glucosa']=bins_PlasmaGlucose
```

```python
df.head()
```

```python
dt=df.loc[:, df.columns != 'PlasmaGlucose']
```

```python
dt.head()
```

```python
columnas=dt.columns
```

```python
anova = []
for col in columnas:      

    a = (stats.f_oneway(
            dt[col][dt['rangos de glucosa']==1],
            dt[col][dt['rangos de glucosa']==2],
            dt[col][dt['rangos de glucosa']==3])) 
    anova.append((col, a))
anova
#si es menor que 0.05 vemos que estan relacionados con la columna bins_quality
columnas_importantes=[i[0] for i in anova if i[1][1]<0.05]
columnas_importantes
```

```python
dt.dtypes
```

```python
columnas=dt.columns
```

```python
data=dt[columnas_importantes]
data.head()
```
