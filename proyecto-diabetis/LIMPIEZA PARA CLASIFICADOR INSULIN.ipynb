```python
import numpy as np
import pandas as pd
import seaborn as sb
from scipy import stats
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('dataset/diabetes.csv')
df.head()
```

```python
df["SerumInsulin"].head()
```

```python
mpg_labels = [1, 2, 3]
cortes=[1,60,140,30000]

bins_insulin = pd.cut(df['SerumInsulin'],cortes, labels=mpg_labels)
bins_insulin.head()
```

```python
bins_insulin.value_counts()
```

```python
df['rangos de insulina']=bins_insulin
```

```python
df.head()
```

```python
dt=df.loc[:, df.columns != 'SerumInsulin']
```

```python
dt.head()
```

```python
columnas=dt.columns
```

```python
anova = []
for col in columnas:      

    a = (stats.f_oneway(
            dt[col][dt['rangos de insulina']==1],
            dt[col][dt['rangos de insulina']==2],
            dt[col][dt['rangos de insulina']==3])) 
    anova.append((col, a))
anova
#si es menor que 0.05 vemos que estan relacionados con la columna bins_quality
columnas_importantes=[i[0] for i in anova if i[1][1]<0.05]
columnas_importantes
```

```python
dt.dtypes
```

```python
columnas=dt.columns
```

```python
data=dt[columnas_importantes]
data.head()
```

```python
data.to_csv("dataset/clasificacion.csv")
```
