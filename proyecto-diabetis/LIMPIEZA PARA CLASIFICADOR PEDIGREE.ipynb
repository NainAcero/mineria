```python
import numpy as np
import pandas as pd
import seaborn as sb
from scipy import stats
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('dataset/diabetes.csv')
df.head()
```

```python
df["DiabetesPedigree"].head()
```

```python
mpg_labels = [0, 1]
cortes=[0,0.50,100]

bins_pedigree = pd.cut(df['DiabetesPedigree'],cortes, labels=mpg_labels)
bins_pedigree.head()
```

```python
bins_pedigree.value_counts()
```

```python
df.isnull().sum()
```

```python
df['rangos padecer diabetes']=bins_pedigree
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
dt = df.loc[:, df.columns != 'DiabetesPedigree']
```

```python
dt.head()
```

```python
dt.isnull().sum()
```

```python
columnas=dt.columns
```

```python
anova = []
for col in columnas:      

    a = (stats.f_oneway(
            dt[col][dt['rangos padecer diabetes']==0],
            dt[col][dt['rangos padecer diabetes']==1])) 
    anova.append((col, a))
anova
#si es menor que 0.05 vemos que estan relacionados con la columna bins_quality
columnas_importantes=[i[0] for i in anova if i[1][1]<0.05]
columnas_importantes
```

```python
columnas=dt.columns
```

```python
data=dt[columnas_importantes]
data.head()
```
